```python
import json
import requests
import time
import datetime

address = "bc1prrjykvc2s928uu2857xzd07qu2sr7xp4pjzafpzah7dd4v4nqkaq2z2e6d"

# Convert the target date to a timestamp
target_date = "2023-12-25"
target_timestamp = int(time.mktime(datetime.datetime.strptime(target_date, "%Y-%m-%d").timetuple()))

# Fetch transaction history
resp = requests.get(f'https://blockchain.info/rawaddr/{address}')
transactions = json.loads(resp.text)['txs']

# Filter transactions by date
filtered_txs = [tx for tx in transactions if tx['time'] >= target_timestamp and tx['time'] < target_timestamp + 86400]

# Fetch UTXOs for filtered transactions
utxos = []
for tx in filtered_txs:
    tx_resp = requests.get(f'https://blockchain.info/rawtx/{tx["hash"]}?format=json')
    tx_data = json.loads(tx_resp.text)
    for output in tx_data['out']:
        if not output['spent']:
            utxos.append(output)

# Print UTXOs
for utxo in utxos:
    try:
        tx_hash = utxo.get('tx_hash', 'N/A')
        tx_output_n = utxo.get('tx_output_n', 'N/A')
        value = utxo.get('value', 0) / 100000000
        print(f"Transaction ID: {tx_hash} - Index: {tx_output_n} - Value: {value} BTC")
    except KeyError as e:
        print(f"Error accessing key in UTXO: {e}")

# Calculate and print total
total = sum(utxo.get('value', 0) for utxo in utxos) / 100000000
print(f"Total BTC available to spend: {total} BTC")
```

```python
import json
import requests
import time
import datetime

address = "1BdQB3qc46odcY67La2ukMV3i2rHJgWpFV"

# Convert the target date to a timestamp
target_date = "2023-12-25"
target_timestamp = int(time.mktime(datetime.datetime.strptime(target_date, "%Y-%m-%d").timetuple()))

# Fetch transaction history
resp = requests.get(f'https://blockchain.info/rawaddr/{address}')
transactions = json.loads(resp.text)['txs']

# Filter transactions by date
filtered_tx_hashes = [tx['hash'] for tx in transactions if tx['time'] >= target_timestamp and tx['time'] < target_timestamp + 86400]

# Fetch UTXOs for the address
utxo_resp = requests.get(f'https://blockchain.info/unspent?active={address}')
utxo_data = json.loads(utxo_resp.text)

# Filter UTXOs by transactions
filtered_utxos = [utxo for utxo in utxo_data.get('unspent_outputs', []) if utxo['tx_hash'] in filtered_tx_hashes]

# Print UTXOs
for utxo in filtered_utxos:
    tx_hash = utxo.get('tx_hash', 'N/A')
    tx_output_n = utxo.get('tx_output_n', 'N/A')
    value = utxo.get('value', 0) / 100000000
    print(f"Transaction ID: {tx_hash} - Index: {tx_output_n} - Value: {value} BTC")

# Calculate and print total
total = sum(utxo.get('value', 0) for utxo in filtered_utxos) / 100000000
print(f"Total BTC available to spend: {total} BTC")
```
